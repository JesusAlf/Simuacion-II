Graficar las salidas obetenidas en la tarea 1

```python
from random import random
import numpy as np
import matplotlib.pyplot as plt 
N=10000000
n=0
lx=[]
ly=[]
for i in range (N):
    x=random()
    y=random()
    if x**2+y**2<1:
        n=n+1
    if i%10000==0 and i>1000:
        estimacion=4*n/(i+1)
        lx.append(i+1)
        ly.append(estimacion)
fig = plt.figure(1, figsize=(8, 8)) 
plt.plot(lx,ly,color="black",markersize=1)
plt.xlabel("Iteraciones")
plt.ylabel("Estimaci√≥n")
plt.grid()
plt.show()
print("El valor de pi simulado es: ",'%3.5f' % (4*n/N))
```

```python
x1 = np.linspace(0, 1, 100)
y1 = np.sqrt(1 - x1**2)               # Ecuaci√≥n del c√≠rculo en el primer cuadrante 
l1=[]
l2= []
for i in range (1000):
    x2 = random()
    y2 = random()
    l1.append(x2)
    l2.append(y2)

fig = plt.figure(1, figsize=(8, 8))        # Aqu√≠ fijamos el formato de la gr√°fica para que sea cuadrada.
plt.plot(x1, y1, color="red", markersize=1) 
plt.scatter(l1, l2, s=20, c="blue")
plt.title("Estimaci√≥n de $\pi$")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
