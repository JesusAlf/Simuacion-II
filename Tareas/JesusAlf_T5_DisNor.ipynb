Elaborar un código en Python para generar una distribución normal por el método de las doce uniformes (Usar la función random( )) 

Hacer el histograma y comparar con el generado por la función gauss( ).

Suponemos que 

$\mu   = 0$

$\sigma^2 = 1$ 



Importamos las librerias a utilizar 

```python
from random import random, gauss 
from matplotlib import pyplot as plt
```

```python
d = []
g = []

for i in range(1000000):
    suma=0
    for j in range(12):
        suma=suma+random()
    d.append(suma-6)
    g.append(gauss(0,1))

plt.figure(figsize = (6, 6))
plt.title("Doce Uniformes")
plt.hist(d, density = 0 , bins = 50, color = "blue", edgecolor="black")

plt.figure(figsize = (6, 6))
plt.title("Distribución Normal")
plt.hist(g, density = 0, bins = 50, color = "red", edgecolor="black")

plt.figure(figsize = (6, 6))
plt.title("Juntas")
plt.hist(g, density = 0, bins = 50, color = "red", edgecolor="black")
plt.hist(d, density = 0, bins = 50, color = "blue", edgecolor="black")
plt.show()

#Por que con la densidad cambia?
#Por que una es mas grande que la otra? 
```
