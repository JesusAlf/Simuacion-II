Elaborar un código en Python para generar una distribución normal por el método de las doce uniformes (Usar la función random( )) 

Hacer el histograma y comparar con el generado por la función gauss( ).

Suponemos que 

$\mu   = 0$

$\sigma^2 = 1$ 



Importamos las librerias a utilizar 

```python
from random import random, gauss 
from matplotlib import pyplot as plt
```

```python
d = []
g = []
n=1000000

for i in range(n):
    suma=0
    for j in range(12):
        suma=suma+random()
        s = suma-6
    d.append(s)
    g.append(gauss(0,1))

plt.figure(figsize = (6, 6))
plt.title("Doce Uniformes")
plt.hist(d, density = 1, bins = 50, color = "blue", edgecolor="black")
plt.show()

plt.figure(figsize = (6, 6))
plt.title("Distribución Normal")
plt.hist(g, density = 1, bins = 50, color = "red", edgecolor="black")
plt.show()
```
